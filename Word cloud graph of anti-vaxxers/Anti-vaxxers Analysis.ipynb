```python
# -*- coding=utf-8 -*-
txt=open("D:\Anti-Vaxx.txt","r",encoding='utf-8').read()
txt = txt.lower()#将英文全部变为小写
```

```python
import nltk
words = nltk.word_tokenize(txt)  #分词
from nltk.stem import SnowballStemmer
snowball_stemmer=SnowballStemmer("english")
print(snowball_stemmer.stem('maximum'))
print(snowball_stemmer.stem('multiply'))
```

```python
freq_dist = nltk.FreqDist(words)
for k,v in freq_dist.items():
    print(str(k)+':'+str(v))

# 把最常见的150个单词拿出来
standard_freq_vector = freq_dist.most_common(150)
print(standard_freq_vector)
```
